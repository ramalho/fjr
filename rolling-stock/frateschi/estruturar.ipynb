```python
import pandas as pd
```

```python
with open('rodante-alfa.txt') as fp:
    lines = fp.readlines()

rows = []
for line in lines:
    line = line.strip()
    if not line:
        continue
    fields = line.split('\t')
    if len(fields) != 4:
        print(repr(line))
        print(fields)
        break
    fields = [f.strip() for f in fields]
    cat = fields[0]
    cod, *rest = reversed(fields[1].split(' - '))
    descr = ' '.join(reversed(rest)).strip()
    qty, price = fields[2:]
    qty = int(qty.split(':')[1])
    price = price.replace('V:', '').replace(';', '').strip()
    rows.append([cat, cod, descr, qty, price])
```

```python
df = pd.DataFrame(rows,
                  columns=['cat', 'cod', 'descr', 'qty', 'price'])
```
